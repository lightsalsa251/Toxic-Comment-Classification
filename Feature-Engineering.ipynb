```python
def lower_clean(text):
    text = str(text)
    text = text.split(' ')
    text = " ".join([w.lower() for w in text])

    return text
```

```python
def punc(text):
    text = re.sub(r'[^A-Za-z\s]',r' ',text)
    text = re.sub(r'\n',r' ',text)    
    
    return text
```

```python
def stop(text):
    #stop_words = set(stopwords.words('english'))
    words = text.split(' ')
    text = " ".join([w for w in words if not w in stop_words])
    return text
```

```python
vec = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='word',
    token_pattern=r'\w{1,}',
    ngram_range=(1, 2),
    max_features=20000)
trn_term_doc = vec.fit_transform(train_features)
test_term_doc = vec.transform(test_features)
```

```python
char_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='char',
    ngram_range=(1, 4),
    max_features=20000)
train_char_features = char_vectorizer.fit_transform(train_features)
test_char_features = char_vectorizer.transform(test_features)
```

```python
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
train_features_sec = hstack((train_char_features, trn_term_doc, up1, count_neg1, ip_count1))
test_features_sec = hstack((test_char_features, test_term_doc, up2, count_neg2, ip_count2))
```

```python
def get_mdl(y):
    y = y.values
    #r = np.log(pr(1,y) / pr(0,y))
    m = LogisticRegression(C=1.2, dual=False)
    #x_nb = x.multiply(r)
    return m.fit(x, y)#, r
```

```python
preds = np.zeros((len(test), len(label_cols)))

for i, j in enumerate(label_cols):
    print('fit', j)
    #m,r = get_mdl(train[j])
    m = get_mdl(train[j])
    preds[:,i] = m.predict_proba(test_x)[:,1]
```
